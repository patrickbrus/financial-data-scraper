# Create Financial Data Scraper
This notebook explores the yfinance Python package and tries to get important financial metrics using the provided functionality.

```python
import yfinance as yf
```

```python
ticker = 'PYPL'
```

```python
# Fetching data for the given ticker
stock = yf.Ticker(ticker)

stock
```

```python
# Getting financials
financials = stock.financials
balance_sheet = stock.balance_sheet
cash_flow = stock.cashflow
```

```python
# Extracting necessary values
# Operating Income or EBIT (Earnings Before Interest and Taxes)
operating_income = financials.loc['Operating Income', :]
```

```python
# Total Assets and Total Liabilities
total_assets = balance_sheet.loc['Total Assets', :]
total_liab = balance_sheet.loc['Total Liabilities', :]
```

```python
# Calculate NOPAT (Net Operating Profit After Taxes)
# Assuming a corporate tax rate - this might need to be adjusted
tax_rate = 0.21  # Example tax rate
nopat = operating_income * (1 - tax_rate)
```
